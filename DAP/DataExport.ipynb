```python
import pandas as pd 
```

# to_csv()

```python
df = pd.read_csv('../Pandas/dataset/groupBy/deliveries.csv')
df.head(2)
```

```python
tempdf = df.groupby('batsman')['batsman_runs'].sum().reset_index()
tempdf.to_csv('batsman_runs_exported.csv',index=False)
```

# to_excel()

```python
tempdf = df.groupby('batsman')['batsman_runs'].sum().reset_index()
tempdf.to_excel('batsman_runs_exported.xlsx',index=False)
tempdf.to_excel('batsman_runs_exported.xlsx',index=False,sheet_name='batsman_runs')
```

```python
tempdf2 = df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum')
```

```python
with pd.ExcelWriter('twosheet.xlsx') as writer: 
    tempdf.to_excel(writer,sheet_name='sheet name 1')
    tempdf2.to_excel(writer,sheet_name='Sheet Name 2')
```

```python
import pandas as pd 
import requests 
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=50637142e9eb92964e364ee4925d5258&language=en-US&page=1')

# Response 200 ka matlab sbhkuch sahi hai 
```

```python
response.json()
```

```python
# mujhe bss  result se mtlb hai usme so json hai so hum bss result lenekeliye we will do this 

response.json()['results']
```

## Converting above stuff into DataFrame 

```python
df= pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average','vote_count']]
```

```python
df.head()
```

#### As we know ki yeah bss 1 page tha so now what we will do here is loop chalayenge page 19 tak & then datafrme banake append krte jayenge 

```python
df = pd.DataFrame()
```

##### appending all data from API till 54 page number 

```python

for i in range(1,54):
     response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=50637142e9eb92964e364ee4925d5258&language=en-US&page={}'.format(i))
     temp_df= pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average','vote_count']]
     df = df.append(temp_df,ignore_index=True)
```

```python
df
```

```python
df.to_csv('movies_gatheredfrom_api.csv')
```

## Types of UnClean Data 

There are 2 kinds of unclean data 

1.) Dirty Data ( Data with Quality issues ):
Dirty Data, also known as low quality data, Low quality data has content issues.

     Duplicated data 
     Missing Data 
     Corrupt Data 
     Inaccurate Data (E.g : height column hai usme negative values hai )

2.) Messy Data (Data with tidiness issues ): Messy data, also known as untidy data,  Untidy data has structural issues. 
Tidy data has the following properties: 

     Each variable forms a column 
     Each observation forms a row 
     Each observational unit forms a table

#  Summary of Data

This is a dataset about 500 patients of which 350 patients participated in a clinical trial.
None of the patients were using Novodra ( a popular injectable insulin) or Auralin ( the oral insulin being researched) as their primary source of insulin before.  All were experiencing elevated HbA1c levels 

All 350 patients were treated with Novodra to establish a baseline HbA1c level and insulin dose. 
After 4 weeks, which isn't enough time to capture all the change in HbA1c that can be attributed by the switch to Auralin or Novodra. 

175 patients switched to Auralin for 24 weeks 
175 patients continued using Novodra for 24 weeks 

Data about patients feeling some adverse effect is also recorded.

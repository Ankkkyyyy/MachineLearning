### Annotations 
Yeah use karke hum apne graph pe ek text ko place karsakte hai 

```python
batters = pd.read_csv('Dataset2/batter.csv')
batters.head(3)
```

```python
# top 100 nikallenge & unsmese random 25 log ko select karenge  
top100batsman = batters.head(100).sample(25,random_state=5)
top100batsman
```

```python
plt.figure(figsize=(15,10))
plt.scatter(top100batsman['avg'],top100batsman['strike_rate'],s=top100batsman['runs'])

for i in range(top100batsman.shape[0]):
    plt.text(top100batsman['avg'].values[i],top100batsman['strike_rate'].values[i],top100batsman['batter'].values[i])
```

### Horizontal & Vertical Lines for Graph 
plt.axhline(130,color='red')

#### Qs. i am the owner of my IPL team i only wnna bif for the batsman whose strike rate is greater than 130 

```python
# i am the owner of my IPL team i only wnna bif for the batsman whose strike rate is greater than 130 
plt.figure(figsize=(15,10))
plt.scatter(top100batsman['avg'],top100batsman['strike_rate'],s=top100batsman['runs'])

plt.axhline(130,color='red')

for i in range(top100batsman.shape[0]):
    plt.text(top100batsman['avg'].values[i],top100batsman['strike_rate'].values[i],top100batsman['batter'].values[i])
```

#### Qs. Findout the batsman whose strike rate is greater than 130 & average is also more than 30 

```python
plt.figure(figsize=(15,10))
plt.scatter(top100batsman['avg'],top100batsman['strike_rate'],s=top100batsman['runs'])

plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')

for i in range(top100batsman.shape[0]):
    plt.text(top100batsman['avg'].values[i],top100batsman['strike_rate'].values[i],top100batsman['batter'].values[i],fontdict={'size':12})
```

## Alternate way of Plotting Matplotlib Graph 

Yeah alternative way isliye helpful hai ki side by side hum multiple graph plot karsakte hai 

```python
# Alternate way of Plotting Matplotlib Graph 

fig, ax  = plt.subplots(figsize=(10,6)) # size bhi yeahi deskte yeah alternative way mai 
ax.scatter(batters['avg'],batters['strike_rate'])
ax.set_title('yo')
ax.set_xlabel('avg')
ax.set_ylabel('strike rate')

# plt.show() ke jageh fig.show()
```

```python
fig ,ax  = plt.subplots(nrows=2,ncols=1)
```

```python
fig ,ax  = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))
ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])


ax[0].set_title('Avg Vs Strike Rate')
# ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs runs')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Runs')
```

```python
fig ,ax  = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])

# ax[0,0].set_title('Avg Vs Strike Rate')
# ax[0].set_xlabel('Average')
# ax[0,1].set_ylabel('Strike Rate')

# ax[1].set_title('Avg Vs runs')
# ax[1].set_xlabel('Average')
# ax[1].set_ylabel('Runs')
```

```python
# 3rd Way 


fig =plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['avg'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['runs'])
```

## Surface Plot 
Yeahi most probably zyaada use hoga at time of Loss Function,  
Loss Function humme yeah batata hai ki , ek Machine Learning Algorithm ka result kitna sahi ya kitna galat hai 

### At ML tbhi yeah use hoga Loss Function ko Plot karte time 


```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy  =np.meshgrid(x,y)
```

```python
z = xx**2+yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# Random stuff
z =np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Contour Plot

3d graph  ko 2d graph krna hoo tbh


```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
p= ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

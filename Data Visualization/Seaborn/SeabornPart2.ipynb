## BoxPlot 

A boxplot is a standardized way of displaying the distribution of data based on five number summary("minimum",first quartile[Q1],median,third Quartile[Q3] and Maximum). 

It can tell you about your outlier and what their values are. 

Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.


```python
# Seaborn provide feature ki dusre categorical plot kesath ushe plot karde 

sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.boxplot(data=tips,x='sex',y='total_bill')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.boxplot(data=tips,y='total_bill')
```

#### regplot-axis level function 
#### Implot - figure level function 

In the simplest invocation , bothfunctions draw a scatterplot of 2 variables x & y and then fit the regression model y ~ x and plot the resulting regression line a 95% confidence interval for that regression 


```python
# Axis level functions
sns.regplot(data=tips,x='total_bill',y='tip') # hue parameter isn't available here 
```

```python
# Figure Level function 
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')  # hue khaali lmplot mai chalega regplot mai nahi 
```

### Plotting Pairwise Relationship  ( PairGrid Vs Pairplot )

Pairplot - automatically smjhjayega kaunsa numeric column hai & kaunsa categorical hai etc 

har combination mai pair pakadke scatter banadeta hai vo 

```python
# sns.load_dataset('iris')
import pandas as pd 
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

Paircolumn internally banta hai uske daddy pairgrid se ,
pair grid se hum kaafi customization krskte basically.

```python
# Pair Grid

g=sns.PairGrid(data=iris,hue='species') 
g.map(sns.scatterplot)
```

```python
# flexibility 

g=sns.PairGrid(data=iris)
# g.map_diag(sns.histplot)
# g.map_offdiag(sns.scatterplot)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)

```

```python
g=sns.PairGrid(data=iris)
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# We can even restrict ki humme kaunse column pe kaam krna hai 

g=sns.PairGrid(data=iris,vars=['sepal_width','petal_length'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

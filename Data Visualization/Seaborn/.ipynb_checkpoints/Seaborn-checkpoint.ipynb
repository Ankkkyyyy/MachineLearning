#### 1.Relational Plots

to see the statistical relation between 2 or more variables 

Bivariate analysis ( 2 column)

Plots :-> Scatterplot, Lineplot 

```python
# Scatter Plot 
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt 
# import plotly.express as px
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

#### Scatter plot in seaborn

```python
# Axis Level Function
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time')
```

```python
# Figure Level Function - Same scatterplot with figure level function - Square shape 
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

Seaborn Library vale pata nahi kyu bolte ki Figure level function zyaada use karo as per their recommendation
isime kind parameter  mai hie sbh change krskte sbh

#### 2. Distribution Plots

used for univariate analysis 

used to find out the distribution 

range of the observation 

central tendency 

is the data bimodal ? yeah bhi batate hai distribution plots

are there outliers? yeah bhi batate hai distribution plots

Plots under distribution plot 

1. histplot 
2. kdeplot 
3. rugplot 


#### Plotting Univariate Histogram

#### Figure Level Function --> displot 

#### Axis Level FUnction --> histplot,kdeplot,rugplot

```python
tips.head(2)
```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
# Alternative using displot 

sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
# bins parameter

sns.displot(data=tips,x='total_bill',kind='hist',bins=20)
```

##### It's also possible to visualize the distribution of a categorical variable using the logic of histogram.
##### Discrete bins are automatically set for categorical variables 

```python

# histogram is acting like a count plot - giving number of counts of each day 

sns.displot(data=tips,x='day',kind='hist')
```

```python
# hue paramter 
# sns.displot(data=tips,x='tip',kind='hist',hue='sex')
# sns.displot(data=tips,x='sex',kind='hist',hue='sex')
# option for clearity -- element='step'
sns.displot(data=tips,x='sex',kind='hist',hue='sex',element='step')
```

```python
# titanic = sns.load_dataset('titanic')
# titanic.head(2)
titanic = pd.read_csv('Titanic-Dataset.csv')
titanic.head(2)
sns.displot(data=titanic,x='Age',kind='hist',element='step',hue='Sex')
```

```python
# faceting using col and row -> this func will not work on histplot function 
sns.displot(data=tips,x='tip',col='sex',kind='hist',hue='sex',element='step')
```

#### kdeplot - Kernel Density Function

Iska maths statistics ke time dekhenge 

```python
tips
sns.kdeplot(data=tips,x='total_bill',hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

#### RugPlot 

Plot marginal distribution by drawing ticks along the x & y axis 

This function is intended to complement other plots by showing the location of individual observations
Yeah dusre graph kesath millke hie iskesath kaam krte hai indiviidually akele nahi 


```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

#### Bivariate Histogram 
A Bivariate histogram bins the data within rectangles  that tile the plot 

and then shows the count of observations within each rectangle with the fill center


```python
sns.histplot(data=tips,x='total_bill',y='tip')
sns.displot(data=tips,x='total_bill',y='tip',kind='hist')
```

#### Bivariate kde plot 

A Bivariate Kde plot smoothes the (x,y) observations with a 2D Gaussian 

```python
sns.kdeplot(data=tips,x='total_bill',y='tip')
```

### Matrix Plot 
HeatMap : Heatmap are rectangular data as color-encoded matrix

ClusterMap 

```python
gap.head(2)
```

```python
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,10)) # this matplotlib type function only works lyk this in axis level function 

# Heatmap is axis level function bss yeahi func se heatmap plot krske aur koi way nahi hai 
sns.heatmap(temp_df)
```

```python
# Graph only for european countries
temp_df= gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
# sns.heatmap(temp_df,annot=True)

# Adding some parameters
sns.heatmap(temp_df,annot=True,linewidth=1.5,cmap='winter')
```

#### Cluster Map  --  hirerchical clustering -- isme internally 1 ml algo kaam krrha jishe kehte hai  Agglomerative hirearchical clustering 

Plot a matrix dataset as a hierarchically - clustered heatmap.

This function requires scipy to be available


```python
iris = px.data.iris()
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
